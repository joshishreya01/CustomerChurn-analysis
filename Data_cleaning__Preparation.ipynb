## Advanced Analytics

### Time Series Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Create a new column 'month_year' for grouping sales
df['month_year'] = df['Date'].dt.to_period('M')

# Group by 'month_year' and sum the 'Total_Cost' to get monthly sales
monthly_sales = df.groupby('month_year')['Total_Cost'].sum()

# Convert PeriodIndex to DateTimeIndex for plotting
monthly_sales.index = monthly_sales.index.to_timestamp()

# Plot Monthly Sales
plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-')

# Formatting x-axis
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=6))

# Labels and title
plt.xlabel('Month-Year')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Over Time')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)  # Adding grid for better readability
plt.tight_layout()

plt.show()

```

### Cohort Segmentation

```python
# Recency: The most recent transaction date per customer
recency = df.groupby('Customer_Name')['Date'].max()

# Frequency: Count of transactions per customer
frequency = df.groupby('Customer_Name')['Date'].count()

# Monetary: Total amount spent per customer
monetary = df.groupby('Customer_Name')['Total_Cost'].sum()

# Combine all three metrics into an RFM DataFrame
rfm = pd.DataFrame({'Recency': recency, 'Frequency': frequency, 'Monetary': monetary})

# Convert Recency to days since last purchase (assuming today is the latest date in the dataset)
latest_date = df['Date'].max()
rfm['Recency'] = (latest_date - rfm['Recency']).dt.days

# Display the RFM DataFrame
print("RFM Analysis:")
print(rfm.head())

# Save the RFM data to a CSV file
rfm.to_csv("Retail_RFM_Analysis.csv", index=True)
print("RFM analysis saved as 'Retail_RFM_Analysis.csv'.")

```

```python
# RFM Analysis
latest_date = df['Date'].max()

rfm = df.groupby('Customer_Name').agg({
    'Date': lambda x: (latest_date - x.max()).days,  # Recency
    'Transaction_ID': 'count',  # Frequency
    'Total_Cost': 'sum'  # Monetary
}).reset_index()

rfm.columns = ['Customer_Name', 'Recency', 'Frequency', 'Monetary']

# Customer Segmentation
def segment_customer(row):
    if row['Recency'] <= 30 and row['Frequency'] >= 15 and row['Monetary'] > 1000:
        return 'P0'
    elif (30 < row['Recency'] <= 90) and (10 < row['Frequency'] <= 15) and (500 < row['Monetary'] <= 1000):
        return 'P1'
    else:
        return 'P2'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)
```

### Churn Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define churn threshold (e.g., 180 days without a transaction)
churn_threshold = 180
latest_date = df['Date'].max()

# Identify churned customers
df['Last_Purchase'] = df.groupby('Customer_Name')['Date'].transform('max')
df['Churn_Status'] = df['Last_Purchase'].apply(lambda x: 'Churned' if (latest_date - x).days > churn_threshold else 'Active')

# Count churned and active customers
churn_counts = df['Churn_Status'].value_counts()

# Plot
plt.figure(figsize=(6, 4))
churn_counts.plot(kind='bar', color=['red', 'green'])
plt.xlabel('Customer Status')
plt.ylabel('Count')
plt.title('Churned vs Active Customers')
plt.xticks(rotation=0)
plt.show()

```

### Analyzing top customers

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure 'Date' is in datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Create a 'month_year' column
df['month_year'] = df['Date'].dt.to_period('M')

# Calculate monetary value per customer (total cost spent)
monetary = df.groupby('Customer_Name')['Total_Cost'].sum()

# Get top 5 customers based on total spend
top_5_customers = monetary.sort_values(ascending=False).head(5).index

# Filter transactions of top 5 customers
top_customers_df = df[df['Customer_Name'].isin(top_5_customers)]

# Aggregate their monthly sales
top_customers_sales = top_customers_df.groupby(['month_year', 'Customer_Name'])['Total_Cost'].sum().unstack(level=1)

# Convert PeriodIndex to DateTimeIndex for plotting
top_customers_sales.index = top_customers_sales.index.to_timestamp()

# Plot
plt.figure(figsize=(12, 6))
top_customers_sales.plot(kind='line', marker='o', ax=plt.gca())

plt.xlabel('Month-Year')
plt.ylabel('Sales (Total Cost)')
plt.title('Monthly Sales of Top 5 Customers')
plt.legend(title='Customer Name')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```
